```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
sns.set(style="whitegrid")
```

```python
df = pd.read_csv("../data/raw/heart_disease.csv")

df.head()
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python
missing = df.isnull().sum().sort_values(ascending=False)
missing
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(df.isnull(), cbar=False)
plt.title("Missing Values Heatmap")
plt.show()
```

```python
df['Heart Disease Status'].value_counts()
sns.countplot(data=df, x='Heart Disease Status')
plt.title("Target Class Distribution")
plt.show()
```

```python
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns

df[numeric_cols].hist(figsize=(15, 12), bins=30)
plt.tight_layout()
plt.show()
```

```python
categorical_cols = df.select_dtypes(include=['object']).columns

for col in categorical_cols:
    plt.figure(figsize=(6,4))
    sns.countplot(data=df, x=col)
    plt.title(f"Distribution of {col}")
    plt.xticks(rotation=45)
    plt.show()
```

```python
plt.figure(figsize=(14,10))
sns.heatmap(df[numeric_cols].corr(), annot=False, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.pairplot(df[numeric_cols[:5]])  # first five numerical features
plt.show()
```
